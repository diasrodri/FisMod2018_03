> Embora as transformações sejam nomeadas em homenagem a Galileu, é o conceito de tempo e espaço absolutos como concebido por Isaac Newton que fornece seu domínio de definição.
    
> Em essência, as transformações de Galileu incorporam a noção intuitiva de adição e subtração de velocidades como vetores.

Seja os referenciais definidos pela figura 1.1: 

- Então as transformações de Galilei serão:

\begin{eqnarray*}
  x^{'} &=&x-vt\\
  y^{'} &=&y\\
  z^{'} &=&z\\
  t^{'} &=&t
\end{eqnarray*}

- Podemos obter as transformações de velovidade por:

\begin{eqnarray*}
  v_x^{'} &=&v_x-v\\
  v_y^{'} &=&v_y\\
  v_z^{'} &=&v_z
\end{eqnarray*}

- Podemos obter as transformações de aceleração por:

\begin{eqnarray*}
  a_x^{'} &=&a_x\\
  a_y^{'} &=&a_y\\
  a_z^{'} &=&a_z
\end{eqnarray*}

Vamos realiza a uma transformação de galileu no problema do oscilador harmonico realizado acima:

### Exemplo: Movimento 1D - Massa Mola

Considere uma partícula de e massa $m=1.0 kg$ pressa a uma mola de constante $K=1.0 N/m$.

![Image1](img/massamola1_1.jpg)

A aceleração para esta partícula será:

\begin{equation}
 a_x(t)=-\frac{k}{m} x(t)
\end{equation}

Logo dado uma uma condição inicial em $t=0$, dada por $v_x(t=0)=1.0$ e $x(t=0)=0.0$ podemos integrar essas equações numericamente e assim obtemos:

```python
k=1.0 #N/m
m=1.0 #Kg
x,vx,ax=0.0,1.2,0.0
xi,vxi,axi,ti=[],[],[],[]
t,dt=0.0,0.001

def rk1(x0,vx0,t0,dt):
    ax=-(k/m)*x0
    x=x0+vx0*dt
    vx=vx0+ax*dt
    return x,vx,ax

for i in range(100000):
    x,vx,ax=rk1(x,vx,t,dt)
    t=t+dt
    if ((i%10)==0):
        xi.append(x)
        vxi.append(vx)
        axi.append(ax)
        ti.append(t)
```

```python
def plotxvxax(t2,x2,vx2,ax2):
    # plot with various axes scales
    plt.figure(num=1, figsize=(12,4), dpi=80, facecolor='w', edgecolor='k')
    
    plt.subplot(131)
    plt.plot(t2,x2,'r-')
    plt.title('Posição x Tempo')
    plt.xlabel('$t(s)$')
    plt.ylabel('$x(m)$')
    plt.grid(True)

    plt.subplot(132)
    plt.plot(t2,vx2,'g-')
    plt.title('Velocidade x Tempo')
    plt.xlabel('$t(s)$')
    plt.ylabel('$vx(m/s)$')
    plt.grid(True)

    plt.subplot(133)
    plt.plot(t2,ax2,'b-')
    plt.title('Aceleração x Tempo')
    plt.xlabel('$t(s)$')
    plt.ylabel('$a_x(m/s^2)$')
    plt.grid(True)

    #plt.gca().yaxis.set_minor_formatter(NullFormatter())
    # Adjust the subplot layout, because the logit one may take more space
    plt.subplots_adjust(top=0.92, bottom=0.08, left=0.10, right=0.95, hspace=0.25,wspace=0.35)
    plt.show()

def plotxvxax_time(t2,x2,vx2,ax2):
    # plot with various axes scales
    plt.figure(num=1, figsize=(12,4), dpi=80, facecolor='w', edgecolor='k')
    
    plt.subplot(131)
    plt.plot(x2,t2,'r-')
    plt.title('Posição x Tempo')
    plt.ylabel('$t(s)$')
    plt.xlabel('$x(m)$')
    plt.grid(True)

    plt.subplot(132)
    plt.plot(vx2,t2,'g-')
    plt.title('Velocidade x Tempo')
    plt.ylabel('$t(s)$')
    plt.xlabel('$vx(m/s)$')
    plt.grid(True)

    plt.subplot(133)
    plt.plot(ax2,t2,'b-')
    plt.title('Aceleração x Tempo')
    plt.ylabel('$t(s)$')
    plt.xlabel('$a_x(m/s^2)$')
    plt.grid(True)

    #plt.gca().yaxis.set_minor_formatter(NullFormatter())
    # Adjust the subplot layout, because the logit one may take more space
    plt.subplots_adjust(top=0.92, bottom=0.08, left=0.10, right=0.95, hspace=0.25,wspace=0.35)
    plt.show()
    
plotxvxax(ti,xi,vxi,axi)
plotxvxax_time(ti,xi,vxi,axi)
```

```python
v=0.1
xlin=[]
vxlin=[]

for i in range(len(xi)):
    xlin.append(xi[i]-v*ti[i])
    vxlin.append(vxi[i]-v)

plotxvxax(ti,xlin,vxlin,axi)
plotxvxax_time(ti,xlin,vxlin,axi)
```
