### Exemplo: Movimento 1D - Massa Mola

Considere uma partícula de e massa $m=1.0 kg$ pressa a uma mola de constante $K=1.0 N/m$.

![Image1](img/massamola1_1.jpg)

A aceleração para esta partícula será:

\begin{equation}
 a_x(t)=-\frac{k}{m} x(t)
\end{equation}

Logo dado uma uma condição inicial em $t=0$, dada por $v_x(t=0)=1.0$ e $x(t=0)=0.0$ podemos integrar essas equações numericamente e assim obtemos:

```python
k=1.0 #N/m
m=1.0 #Kg
x,vx,ax=0.0,-1.2,0.0
xi,vxi,axi,ti=[],[],[],[]
t,dt=0.0,0.001

def acel(k,m,x0):
#    ax=-(k/m)*x0
    ax=10.0
    return ax

def rk1(x0,vx0,t0,dt):
    ax=acel(k,m,x0)
    x=x0+vx0*dt
    vx=vx0+ax*dt
    return x,vx,ax

for i in range(1000):
    x,vx,ax=rk1(x,vx,t,dt)
    t=t+dt
    if ((i%10)==0):
        xi.append(x)
        vxi.append(vx)
        axi.append(ax)
        ti.append(t)
```

```python
def plotxvxax(t2,x2,vx2,ax2):
    # plot with various axes scales
    plt.figure(num=1, figsize=(12,4), dpi=80, facecolor='w', edgecolor='k')
    
    plt.subplot(131)
    plt.plot(t2,x2,'r-')
    plt.title('Posição x Tempo')
    plt.xlabel('$t(s)$')
    plt.ylabel('$x(m)$')
    plt.grid(True)

    plt.subplot(132)
    plt.plot(t2,vx2,'g-')
    plt.title('Velocidade x Tempo')
    plt.xlabel('$t(s)$')
    plt.ylabel('$vx(m/s)$')
    plt.grid(True)

    plt.subplot(133)
    plt.plot(t2,ax2,'b-')
    plt.title('Aceleração x Tempo')
    plt.xlabel('$t(s)$')
    plt.ylabel('$a_x(m/s^2)$')
    plt.grid(True)

    #plt.gca().yaxis.set_minor_formatter(NullFormatter())
    # Adjust the subplot layout, because the logit one may take more space
    plt.subplots_adjust(top=0.92, bottom=0.08, left=0.10, right=0.95, hspace=0.25,wspace=0.35)
    plt.show()

def plotxvxax_time(t2,x2,vx2,ax2):
    # plot with various axes scales
    plt.figure(num=1, figsize=(12,4), dpi=80, facecolor='w', edgecolor='k')
    
    plt.subplot(131)
    plt.plot(x2,t2,'r-')
    plt.title('Posição x Tempo')
    plt.ylabel('$t(s)$')
    plt.xlabel('$x(m)$')
    plt.grid(True)

    plt.subplot(132)
    plt.plot(vx2,t2,'g-')
    plt.title('Velocidade x Tempo')
    plt.ylabel('$t(s)$')
    plt.xlabel('$vx(m/s)$')
    plt.grid(True)

    plt.subplot(133)
    plt.plot(ax2,t2,'b-')
    plt.title('Aceleração x Tempo')
    plt.ylabel('$t(s)$')
    plt.xlabel('$a_x(m/s^2)$')
    plt.grid(True)

    #plt.gca().yaxis.set_minor_formatter(NullFormatter())
    # Adjust the subplot layout, because the logit one may take more space
    plt.subplots_adjust(top=0.92, bottom=0.08, left=0.10, right=0.95, hspace=0.25,wspace=0.35)
    plt.show()
    
plotxvxax(ti,xi,vxi,axi)
plotxvxax_time(ti,xi,vxi,axi)
```

## 1.7 Postulados da Teoria da Relatividade Especial


Enunciados com simplicidade, esses postulados são:

**Postulado 1:** As leis da Física são as mesmas em qualquer referencial inercial. Não existe nenhum referencial inercial privilegiado.

### 1.8.2 - Transformações de Lorentz.

Seja um evento ocorrendo num referencial inercial $S$, caracterizado por coordenadas espaço-tempo $(x,y,z,t)$.

Em um segundo referencial $S^{'}$, o mesmo evento é caracterizado por coordendas espaço-tempo $(x^{'},y^{'},z^{'},t^{'})$. 

Deseja-se encontra realções funcionais do tipo:
 
$$x^{'}=f(x,y,z,t)~~;~~y^{'}=f(x,y,z,t)~~;~~z^{'}=f(x,y,z,t)~~;~~t^{'}=f(x,y,z,t),$$

isto é, equações de transformação que relacionam as coodenadas espaço-tempo de um envento do ponto de vista de um observador inercial, com as coordenadas do mesmo evento do ponto de vista de outro observador inercial.

Juntamente com os postulados fundamentasis da teoria da relatividade, deve-se recorrer a uma hipótese de que **o espaço e o tempo são homogêneos**, isto é , **todos os pontos no espaço e tempo são equivalentes**.

Essa hipótese exige que as equações de transformação senam **lineares**, tal que a forma mais geral que podem assumir é:

\begin{eqnarray*}
  x^{'} &=&a_{11}x+a_{12}y+a_{13}z+a_{14}t\\
  y^{'} &=&a_{21}x+a_{22}y+a_{23}z+a_{24}t\\
  z^{'} &=&a_{31}x+a_{32}y+a_{33}z+a_{34}t\\
  t^{'} &=&a_{41}x+a_{42}y+a_{43}z+a_{44}t ~~~~~~~~~~ (1.29)
\end{eqnarray*}

Uma outro hipótese que se pode admitir para o caso partícular do espaço é que ele é **isotrópico**, isto é, **todas as direções no espaço são equivalentes**. 

Essa hipótese permite simplificar o problema supondo que a velocidade relatica $v$ dos referenciais $S$ e $S^{'}$ ocorra paralelamente aos eixos comuns $x$ e $x^{']$, e que os planos correspondentes sejam mantidos paralelos.

Por exemplo se $v=0$, os dois referenciais coincidem o tempo topo tal que $a_{ii}=1$, com $i=1,2,3,4$ e  $a_{ij}=0$, com $i \neq j=1,2,3,4$

Queremos encontrar quais são os coeficiente para o caso geral $v\neq 0$.

O eixo $x$ coincidirá continuamente com o eixo $x^{'}$, se pontos sobre o eixo $x(y=z=0)$ resultarem em pontos sobre o eixo $x^{'}(y^{'}=z^{'}=0)$. Assim de $(1.29)$,

\begin{eqnarray*}
  0 &=&a_{21}x+a_{22}0+a_{23}0+a_{24}t=a_{21}x+a_{24}t\\
  0 &=&a_{31}x+a_{32}0+a_{33}0+a_{34}t=a_{31}x+a_{34}t
\end{eqnarray*}

Para essa igualdade ser satisféita temos que, $a_{21}=a_{24}=a_{31}=a_{34}=0$ e assim,

\begin{eqnarray*}
  y^{'} &=&a_{22}y+a_{23}z\\
  z^{'} &=&a_{32}y+a_{33}z.
\end{eqnarray*}

Para que os pontos do plano $xz(y=0)$ e $xy(z=0)$ resultem em pontos do plano  $x^{'}z^{'}(y^{'}=0)$ e  $x^{'}y^{'}(z^{'}=0)$, respectivamente temos,

\begin{eqnarray*}
  0^{'} &=&a_{22}0+a_{23}z=a_{23}z\\
  0^{'} &=&a_{32}0+a_{33}z=a_{33}z,
\end{eqnarray*}

é necessário que $a_{23}=a_{33}=0$  e com obtemos,

\begin{eqnarray*}
  y^{'} &=&a_{22}y\\
  z^{'} &=&a_{33}z.
\end{eqnarray*}

 Como a medida de uma barra de tamanho unitário $y=1m(z=1m)$ visto de um referencial $S$ deve ser o mesmo quando visto por outro observador em $S^{'}$ tal que $y^{'}=1m(z^{'}=1m)$ assim como vemos na figura (1.12), temos das equações acima que $a_{22}=a_{33}=1$. Logo obtemos,
 
 \begin{eqnarray*}
  y^{'} &=&y\\
  z^{'} &=&z.
\end{eqnarray*}
 
<img src="https://github.com/diasrodri/FisMod2018_03/blob/img/Fig1.12.png?raw=true" width="500">

Restam ainda analizar as equações de transformação para $x^{'}$ e $t^{'}$. **Não de pode esperar** a dependência de $t^{'}$ com $y$ e $z$, pois relógios colocados simetricamente nestes eixos permendiculares ao movimento, por exemplo $y_1=1$, $z_1=1$ e $y_2=-1$, $z_2=-1$ discordariam quando vistos de $S^{'}$. Podemos ver isso de,

\begin{eqnarray*}
  t_1^{'} &=&a_{41}x+a_{42}1+a_{43}1+a_{44}t\\
  t_2^{'} &=&a_{41}x-a_{42}1-a_{43}1+a_{44}t
\end{eqnarray*}

daí terimos que, $$t_1^{'}-t_2^{'}=2a_{42}+2a_{43}$$.

Lembrando que nessas direções não exite movimentos relativos e as leituras dos relógios devem ser as mesmas $(t_1^{'}-t_2^{'}=0)$.

Para que isso ocorra, é necessário que $a_{42}=a_{43}=0$ e assim, 

\begin{eqnarray*}
  t^{'} &=&a_{41}x+a_{44}t
\end{eqnarray*}

Quanto a transformação para  $x^{'}$, recorre-se a Fig 1.13, onde um ponto sobre o eixo  $x^{'}=0$ em $S^{'}$ se desloca com velocidade $v$ em relação ao referencial $S$.

<img src="https://github.com/diasrodri/FisMod2018_03/blob/img/Fig1.13.png?raw=true" width="500">


Para que se tenha $x^{'}=y^{'}=z^{'}=0$ quando $x=vt$ e $y=z=0$ temos,

\begin{eqnarray*}
  0 &=&a_{11}vt+a_{12}0+a_{13}0+a_{14}t=a_{11}vt+a_{14}t
\end{eqnarray*}

obtemos que $a_{14}=-a_{11}v$

Como essa equação de ser sempre válida qualquer que sejam os pontos $y=z$ temos que $a_{12}=a_{13}=0$ finaliznaod que,

\begin{eqnarray*}
  x^{'} &=&a_{11}(x-vt)\\
  t^{'} &=&a_{41}x+a_{44}t
\end{eqnarray*}

Por fím, é necessário determinar os três coeficientes $a_{11}$, $a_{41}$ e $a_{44}$.

Isso pode ser feito a partir do $2^o$ Postulado de Einstein.

Suponha que em $t=0$, uma onda eletromagnética esférica seja emitida da origem de $S$, coincidente com a oreigem de $S^{'}$ no instante $t^{'}=0$.

A onda se propaga em todas as direções com velocidade $c$ em ambos os referenciais. 

A frente de onda é descrita pela equação de uma uma esfera cujo o raio se expande no tempo com velocidade $c$ tal que, para cada referencial,

\begin{eqnarray*}
  x^2+y^2+z^2 &=& c^2 t^2\\
  x^{'2}+y^{'2}+z^{'2} &=& c^2 t^{'2}
\end{eqnarray*}

Substituindo as equações acima temos,

\begin{eqnarray*}
  x^2+y^2+z^2 &=& c^2 t^2\\
  [a_{11}(x-vt)]^{2}+y^{2}+z^{2} &=& c^2 (a_{41}x+a_{44}t)^{2}
\end{eqnarray*}

\begin{eqnarray*}
  x^2 &=& c^2 t^2\\
  a_{11}^{2}(x^{2}+v^{2}t^{2}-2xvt)&=& c^2 (a_{41}^{2}x^{2}+a_{44}^{2}t^{2}+2 a_{41} x a_{44} t)^{2}
\end{eqnarray*}


\begin{eqnarray*}
  x^2 &=& c^2 t^2\\
  (a_{11}^{2}-c^2 a_{41}^{2})x^{2} -2( a_{11}^{2} v + c^2 a_{41} a_{44} ) xt &=&  (c^2 a_{44}^{2}-v^2 a_{11}^{2}) t^{2}
\end{eqnarray*}


Comparando essas equações vemos que:

\begin{eqnarray*}
  a_{11}^{2}-c^2 a_{41}^{2} &=& 1\\
  v a_{11}^{2} + c^2 a_{41} a_{44}  &=& 0 \\
  c^2 a_{44}^{2}-v^2 a_{11}^{2} &=& c^{2}
\end{eqnarray*}

Da segunda equação temos,

\begin{eqnarray*}
 v^2 a_{11}^{2}    &=& - v c^2 a_{41} a_{44} 
\end{eqnarray*}

Substituindo na terceira e na primeira,

\begin{eqnarray*}
  c^2 a_{44}^{2}+ v c^2 a_{41} a_{44} = c^{2} & \Rightarrow & a_{44}+ v a_{41} &=& 1 \\
   -c^2 a_{41} a_{44}/v -c^2 a_{41}^{2} = 1 & \Rightarrow & -c^2(a_{44}/v +a_{41}) = 1
\end{eqnarray*}

Podemos verificar que a solução do sistema de equações é :

\begin{eqnarray*}
  a_{11} &=& \gamma\\
  a_{44} &=& \gamma\\
  a_{41} &=& -\gamma v/c^2 
\end{eqnarray*}

onde 

\begin{eqnarray*}
  \gamma &=& \frac{1}{\sqrt{1-\beta^2}}\\
  \beta &=& v/c
\end{eqnarray*}

Substituindo esses coeficientes obtemos as **equações de transformações de Lorentz**:


\begin{eqnarray*}
  x^{'} &=&\gamma(x-vt)\\
  y^{'} &=&y\\
  z^{'} &=&z\\
  t^{'} &=&\gamma(t-vx/c^2) ~~~~~~~~~~ (1.39)
\end{eqnarray*}

Se as coordenadas de espaço-tempo fossem observadas apartir de $S^{'}$, ao invés de S, a única mudança permitida pelo princípio da relatividade é no sentido da velocidade relativa ir de $v$ para $-v$. Assim, $x\Rightarrow x^{'}$ , $y\Rightarrow y^{'}$, $z\Rightarrow z^{'}$ e $t\Rightarrow t^{'}$ e assim obtermos as **transformações de Lorentz inversas**:

\begin{eqnarray*}
  x &=&\gamma(x^{'}+vt^{'})\\
  y &=&y^{'}\\
  z &=&z^{'}\\
  t &=&\gamma(t^{'}+vx^{'}/c^2) ~~~~~~~~~~ (1.39)
\end{eqnarray*}

### 1.8.3 - Transformações de Velocidade e Aceleração

Vamos obter inicialmente as transformações de velocidade. Para tal, considere dois eventos consecutivos $A(x_1,y_1,z_1,t_1)$ e $B(x_2,y_2,z_2,t_2)$, assim as variações no espaço-tempo serão no referencial $S^{'}$:

\begin{eqnarray*}
  x_2^{'} -x_1^{'} &=&\gamma[ (x_2-x_1) -v(t_2-t_1) ]\\
  y_2^{'} -y_1^{'} &=&(y_2-y_1) \\
  z_2^{'} -z_1^{'} &=&(z_2-z_1) \\
  t_2^{'} -t_1^{'} &=&\gamma[ (t_2-t_1) -v(x_2-x_1)/c^2 ]
\end{eqnarray*}

Essas equações assumem o limite infinitesimal por:

\begin{eqnarray*}
  dx^{'} &=&\gamma(dx-vdt)\\
  dy^{'} &=&y\\
  dz^{'} &=&z\\
  dt^{'} &=&\gamma(dt-vdx/c^2)
\end{eqnarray*}

Notando que 

\begin{eqnarray*}
  u_x^{'}=\frac{dx^{'}}{dt^{'}} &=&\frac{\gamma(dx-vdt)}{\gamma(dt-vdx/c^2)}=\frac{(dx/dt-v)}{(1-(dx/dt)v/c^2)}\\
  u_y^{'}=\frac{dy^{'}}{dt^{'}} &=&\frac{dy}{\gamma(dt-vdx/c^2)}=\frac{(dy/dt)}{\gamma(1-(dx/dt)v/c^2)}\\
  u_z^{'}=\frac{dz^{'}}{dt^{'}} &=&\frac{dz}{\gamma(dt-vdx/c^2)}=\frac{(dz/dt)}{\gamma(1-(dx/dt)v/c^2)}
\end{eqnarray*}

obtemos então:

\begin{eqnarray*}
  u_x^{'}=&=&\frac{(u_x-v)}{(1-vu_x/c^2)}\\
  u_y^{'}=&=&\frac{u_y}{\gamma(1-u_xv/c^2)}\\
  u_z^{'}=&=&\frac{u_z}{\gamma(1-u_xv/c^2)}
\end{eqnarray*}

Com transformações inversas dada por:

\begin{eqnarray*}
  u_x=&=&\frac{(u_x^{'}+v)}{(1+u_x^{'}v/c^2)}\\
  u_y=&=&\frac{u_y^{'}}{\gamma(1+u_x^{'}v/c^2)}\\
  u_z=&=&\frac{u_z^{'}}{\gamma(1+u_x^{'}v/c^2)}
\end{eqnarray*}
