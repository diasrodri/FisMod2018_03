## 2.3 Teoria Clássica de Rayleigh-Jeans para a Radiação de Corpo Negro

Relação entre a radiância espectral de corpo negro e a densidade de energia da cavidade é obtida sabendo que a energia por unidade de volume da radiação com freqüência entre  $f$ e $f +df$ será dada por,
$$\frac{dU}{dV}(f)=u_{f}(f)$$

Uma vez que a cavidade está em equilíbrio termico, podemos considerar que a radiação de cavidade é isotrópica.

<img src="https://github.com/diasrodri/FisMod2018_03/blob/img/Fig2.n.png?raw=true" width="500">


A parte da energia, descrita acima, que se propaga em direção à parede da cavidade, vinda de direções definidas pelos intervalos de ângulos esféricos $(\theta, \theta+d\theta)$ e $(\phi, \phi+d\phi)$ é

$$\frac{dU}{dV}(f,\theta,\phi)=\frac{u_{f}(f)\sin\theta d\theta d\phi}{4\pi} $$

A seção transversal do feixe de radiação vindo desta direção e passando pelo buraco de área $A$ é $A\cos\theta$.

A radiação se propaga com a velocidade da luz c e desloca uma distância $dx=c dt$ durante o intervalo de tempo $dt$.

O volume de radiação que passa pelo buraco durante o intervalo de tempo $dt$ é $dV=(A\cos\theta)(cdt)$.

A energia desta radiação é,
$$dU(f,\theta,\phi)=\frac{A c u_{f}(f)\sin\theta\cos\theta dt d\theta d\phi}{4\pi} $$.

Radiação saindo de um buraco numa cavidade, a **radiância espectral** é obtida integrando sobre todos os ângulos e dividindo pela área da superfície emissora (o buraco) e dividindo pelo intervalo de tempo. Logo,

$$I_f(f)=\frac{1}{A}\frac{dU(f)}{dt}=\int_{0}^{\pi/2}\int_{0}^{2\pi}\frac{c u_{f}(f) \sin\theta\cos\theta d\theta d\phi}{4\pi} $$

$$I_f(f)=\frac{c u_{f}(f)}{4\pi}\int_{0}^{\pi/2}\sin\theta\cos\theta d\theta \int_{0}^{2\pi}d\phi$$

$$I_f(f)=\frac{c u_{f}(f)}{2}\int_{0}^{\pi/2}\sin\theta\cos\theta d\theta $$

$$I_f(f)=\frac{c u_{f}(f)}{4}$$


A densidade de energia espectral da cavidade será dada pela energia média $<E>$ de uma onda eletromagnética de freqüência $f$ multiplicada pelo número $n_f(f)$ de ondas por unidade de volume, com frequência no intervalo $(f,f+df)$.

$$u_{f}(f)=<E>n_f(f)$$

A **radiância espectral** será então,

$$I_f(f)=\frac{c}{4} <E> n_f(f)$$


>Energia média de uma onda será dada pelo **Teorema de equipartição da energia**
>
>Num sistema em equilíbrio térmico a uma temperatura $T$, a energia cinética média por grau de liberdade vale $<K>=\frac{1}{2}k_b T$.
>
>Uma onda eletromagnética num dado estado de polarização linear é análoga a um oscilador harmônico unidimensional(MHS).
>
>Para um sistema que seja MHS, a energia potencial média (sobre um período) é igual à energia cinética média: $<K>=<U>$.
>
>Portanto, a energia total média é
> $$<E>=<K>+<U>=2<K>$$
> $$<E>=k_b T$$<br>
>Na teoria clássica, a energia média é independente da freqüência.

O vetor de onda $\vec{k}$ para um cavidade quadrada de lados $L$ será $\vec{k}=\frac{\pi}{L}(q_x\hat{e}_x+q_y\hat{e}_y+q_z\hat{e}_z)$, onde $q_{x,y,z}=1,2,3,...$

<img src="https://github.com/diasrodri/FisMod2018_03/blob/img/Fig2.n1.png?raw=true" width="400">


No espaço do vetor $k$, estes valores definem pontos formando uma rede cúbica de intervalo unitário.

O número de pontos numa dada região é igual ao volume desta região.

O número de ondas no intervalo de freqüência $df$ é dado pelo número de pontos num oitavo de casca esférica de raio $q$ e espessura $dq$.

Portanto, é dado pelo volume do oitavo de casca: 
$$N_f(f)df=\frac{1}{8} 4\pi q^2 dq $$

O modulo de $q$ é $q=\sqrt{q_x^2+q_y^2+q_z^2}=\frac{L}{\pi}k$ e assim $dq=\frac{L}{\pi}dk$ 
$$N_f(f)=\frac{1}{8} 4\pi (\frac{L}{\pi}k)^2 (\frac{L}{\pi}) dk = \frac{L^3}{2\pi^2}   k^2 dk $$

Como $\omega=2\pi f = c k$ temos que $dk=\frac{2\pi}{c}df$ e assim,
$$N_f(f)=\frac{L^3}{2\pi^2} \left(\frac{2\pi f}{c}\right)^2 \frac{2\pi}{c}df $$
$$N_f(f)df=  \frac{4\pi L^3}{c^3}f^2 df $$

Multiplicando o resultado acima por 2 devido ao fato de existir dois estados distintos de polarização, e dividindo pelo volume $V=L^3$ obtemos:

$$n_f(f)=\frac{N_f(f)}{L^3}= \frac{8\pi}{c^3}f^2 $$


e portanto a **radiância espectral** será:

$$I_f(f)=\frac{c}{4} k_bT \frac{8\pi}{c^3}f^2$$

$$I_f(f)=k_bT \frac{2\pi}{c^2}f^2$$ 


No limite de pequenos valores de $f$, a teoria clássica fornece resultados adequados.

No entanto, o resultado clássico diverge cada vez mais dos dados experimentais à medida que a freqüência cresce. 

Não apresenta o máximo presente nos dados experimentais.

E também é físicamente absurdo, pois implicaria numa potência total emitida infinita pois a integral da radiância sobre todas as freqüências e sobre a superfície do corpo diverge para infitito.

A teoria clássica revela-se não apenas incapaz de reproduzir a fenomenologia, mas intrinsecamente paradoxal.

Como a patologia está na região de altas freqüências, fala-se da catástrofe do ultravioleta.
